```python
import openpyxl

# Load the Excel file
workbook = openpyxl.load_workbook('book1.xlsx')

# Get the sheet that contains the table
sheet = workbook['customers']

# Get the table by name
table = sheet.tables['customers_table']

table_range = table.ref

# Get the data from the table
data = []
for row in sheet[table_range]:
    data.append([cell.value for cell in row])

# Print the data
for row in data:
    print(row)
```
